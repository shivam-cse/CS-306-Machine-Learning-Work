## Question No - 2

```python
data = df.drop('id', axis=1)

ratio = 0.1
for i in range(9):
  shuffle_data = data.sample(frac=1)
  x_len = int(ratio * len(data))
  print("Splitting Data Into To Part -> ", int(ratio * 100)  , ":", int(100 - ratio * 100))

  x_data = shuffle_data[ : x_len]
  y_data = shuffle_data[x_len : ]
  print(x_data)
  print(y_data)
  ratio += 0.1
```

## **Question No -3**

```python
data = df.drop('id', axis=1)
data
print("row - wise mean : ")
print(data.mean(axis=1))

print("\nmean columns-wise : ")
print(data.mean(axis=0))

print("\nmean over all :", data.stack().mean())
```

```python
print(data)
print("row - wise median : ")
print(data.median(axis=1))

print("\nmedian columns-wise : ")
print(data.median(axis=0))

print("\nmedian over all :", data.stack().median())
```

```python
print("row - wise mode : ")
print(data.mode(axis=1))

print("\nmode columns-wise : ")
print(data.mode(axis=0))

print("\nmode over all :", data.stack().mode())
```

```python
print("row - wise sd : ")
print(data.std(axis=1))

print("\nsd columns-wise : ")
print(data.std(axis=0))

print("\n std over all :", data.stack().std())
```

```python
data_scores = pd.read_csv('student_scores.csv')
```

```python
import matplotlib.pyplot as plt
x = data_scores.drop('Scores', axis=1)
y = data_scores['Scores']
x = x.sort_values(by=['Hours'])
y = y.sort_values()
plt.scatter(x,y)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Hours v/s  Scores for students')
plt.plot(x,y)
```
